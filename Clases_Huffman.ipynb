```python
import requests
url = "https://raw.githubusercontent.com/Mateo17pi/Poema/main/poema.txt"

mjs = requests.get(url).text

class Nodes:
    def __init__(self, probabilidad, simbolo, izquierda = None, derecha = None):

        self.probabilidad = probabilidad

        self.simbolo = simbolo

        self.izquierda = izquierda

        self.derecha = derecha

        self.codigo = ''

def CalculadoraProbabilidad(datos):
    simbolos = dict()
    for n in datos:
        if simbolos.get(n) == None:
            simbolos[n] = 1
        else:
            simbolos[n] += 1
    return simbolos

codigos = dict()

def CalculadoraCodigos(nodo, valor = ''):
    Nuevovalor = valor + str(nodo.codigo)

    if(nodo.izquierda):
        CalculadoraCodigos(nodo.izquierda, Nuevovalor)
    if(nodo.derecha):
        CalculadoraCodigos(nodo.derecha, Nuevovalor)

    if(not nodo.izquierda and not nodo.derecha):
        codigos[nodo.simbolo] = Nuevovalor

    return codigos

def SalidaCodificada(datos, codificacion):
    salida = []
    for i in datos:
        salida.append(codificacion[i])

    cadena = ''.join([str(item) for item in salida])
    return cadena

def TotalGain(datos, codificacion):
    Antesdecomprimir = len(datos) * 8
    Despuesdecomprimir = 0
    simbolos = codificacion.keys()
    for simbolo in simbolos:
        contar = datos.count(simbolo)
        Despuesdecomprimir += contar * len(codificacion[simbolo])
    print("Espacio usado antes de comprimir en bits:", Antesdecomprimir)
    print("Espacio usado despues de comprimir en bits:",  Despuesdecomprimir)

def HuffmanEncoding(datos):
    simboloW = CalculadoraProbabilidad(datos)
    simbolos = simboloW.keys()
    probabilidades = simboloW.values()
    print("Simbolos: ", simbolos)
    print("Probabilidades: ", probabilidades)

    nodos = []

    for simbolo in simbolos:
        nodos.append(Nodes(simboloW.get(simbolo), simbolo))

    while len(nodos) > 1:
        nodos = sorted(nodos, key = lambda x: x.probabilidad)

        derecha = nodos[0]
        izquierda = nodos[1]

        izquierda.codigo = 0
        derecha.codigo = 1

        nuevoNodo = Nodes(izquierda.probabilidad + derecha.probabilidad, izquierda.simbolo + derecha.simbolo, izquierda, derecha)

        nodos.remove(izquierda)
        nodos.remove(derecha)
        nodos.append(nuevoNodo)

    huffmanEncoding = CalculadoraCodigos(nodos[0])
    print("Simbolos y su codigo", huffmanEncoding)
    TotalGain(datos, huffmanEncoding)
    salidaCodificada = SalidaCodificada(datos,huffmanEncoding)
    return salidaCodificada, nodos[0]



datos = mjs
print(datos)
codificacion, ArbolH = HuffmanEncoding(datos)
print("Salida codificada", codificacion)
```

```python
def HuffmanDecodificacion(Datoscodificados, ArbolH):
    Cabeza = ArbolH
    Salidadecodificada = []
    for x in Datoscodificados:
        if x == '1':
            ArbolH = ArbolH.derecha
        elif x == '0':
            ArbolH = ArbolH.izquierda
        try:
            if ArbolH.izquierda.simbolo == None and ArbolH.derecha.simbolo == None:
                pass
        except AttributeError:
            Salidadecodificada.append(ArbolH.simbolo)
            ArbolH = Cabeza

    string = ''.join([str(item) for item in Salidadecodificada])
    return string

print("Decoded Output", HuffmanDecodificacion(codificacion, ArbolH))
```
